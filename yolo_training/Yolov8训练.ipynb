```python
# è®¾ç½®å‚æ•°
DATASET_PATH = "my_dataset_2"  # æ•°æ®é›†è·¯å¾„ï¼Œè¯·æ ¹æ®å®é™…æƒ…å†µä¿®æ”¹
PROJECT_NAME = "yolov8_training"   # é¡¹ç›®åç§°
MODEL_TYPE = "yolov8s.pt"          # æ¨¡å‹ç±»å‹: yolov8n.pt, yolov8s.pt, yolov8m.pt, yolov8l.pt, yolov8x.pt
EPOCHS = 100                       # è®­ç»ƒè½®æ•°
BATCH_SIZE = 16                    # æ‰¹æ¬¡å¤§å°
IMAGE_SIZE = 640                   # å›¾åƒå°ºå¯¸
TRAIN_RATIO = 0.8                  # è®­ç»ƒé›†æ¯”ä¾‹
VAL_RATIO = 0.2                    # éªŒè¯é›†æ¯”ä¾‹
CONF_THRESHOLD = 0.05              # é™ä½æ£€æµ‹ç½®ä¿¡åº¦é˜ˆå€¼ï¼Œæé«˜æ£€å‡ºç‡

# å®‰è£…å¿…è¦çš„åº“
!pip install ultralytics

# å¯¼å…¥å¿…è¦çš„åº“
import os
import yaml
import json
import random
import shutil
import numpy as np
from ultralytics import YOLO
from google.colab import drive
from IPython.display import display, Image

# è®¾ç½®éšæœºç§å­ä»¥ç¡®ä¿å¯é‡å¤æ€§
random.seed(42)
np.random.seed(42)

# åˆ›å»ºé¡¹ç›®ç›®å½•
os.makedirs(PROJECT_NAME, exist_ok=True)

# æ£€æŸ¥æ•°æ®é›†è·¯å¾„æ˜¯å¦å­˜åœ¨
if not os.path.exists(DATASET_PATH):
    raise FileNotFoundError(f"æ•°æ®é›†è·¯å¾„ {DATASET_PATH} ä¸å­˜åœ¨ï¼Œè¯·æ£€æŸ¥è·¯å¾„æ˜¯å¦æ­£ç¡®")

# æ£€æŸ¥æ•°æ®é›†ç»“æ„
images_path = os.path.join(DATASET_PATH, 'images/')
labels_path = os.path.join(DATASET_PATH, 'labels/')
classes_path = os.path.join(DATASET_PATH, 'classes.txt')

if not os.path.exists(images_path):
    raise FileNotFoundError(f"å›¾åƒè·¯å¾„ {images_path} ä¸å­˜åœ¨ï¼Œè¯·ç¡®ä¿æ•°æ®é›†åŒ…å« images æ–‡ä»¶å¤¹")
if not os.path.exists(labels_path):
    raise FileNotFoundError(f"æ ‡ç­¾è·¯å¾„ {labels_path} ä¸å­˜åœ¨ï¼Œè¯·ç¡®ä¿æ•°æ®é›†åŒ…å« labels æ–‡ä»¶å¤¹")

# è¯»å–ç±»åˆ«ä¿¡æ¯
class_names = []
if os.path.exists(classes_path):
    with open(classes_path, 'r') as f:
        class_names = [line.strip() for line in f.readlines()]
    print(f"ä» classes.txt ä¸­è¯»å–åˆ° {len(class_names)} ä¸ªç±»åˆ«: {class_names}")
else:
    # å¦‚æœæ²¡æœ‰æ‰¾åˆ° classes.txtï¼Œå°è¯•ä»æ ‡ç­¾æ–‡ä»¶ä¸­æ¨æ–­
    print("æœªæ‰¾åˆ° classes.txt æ–‡ä»¶ï¼Œå°è¯•ä»æ ‡ç­¾æ–‡ä»¶ä¸­æ¨æ–­ç±»åˆ«...")
    max_class_id = -1
    for label_file in os.listdir(labels_path):
        if label_file.endswith('.txt'):
            with open(os.path.join(labels_path, label_file), 'r') as f:
                for line in f:
                    parts = line.strip().split()
                    if parts:
                        class_id = int(parts[0])
                        max_class_id = max(max_class_id, class_id)

    # åˆ›å»ºé»˜è®¤ç±»åˆ«åç§°
    if max_class_id >= 0:
        class_names = [f"class_{i}" for i in range(max_class_id + 1)]
        print(f"ä»æ ‡ç­¾æ–‡ä»¶ä¸­æ¨æ–­å‡º {len(class_names)} ä¸ªç±»åˆ«")
    else:
        print("æ— æ³•ç¡®å®šç±»åˆ«æ•°é‡ï¼Œå°†ä½¿ç”¨é»˜è®¤ç±»åˆ«åç§°")
        class_names = ["object"]

# è¯»å– note.jsonï¼ˆå¦‚æœå­˜åœ¨ï¼‰
note_info = {}
note_path = os.path.join(DATASET_PATH, 'note.json')
if os.path.exists(note_path):
    try:
        with open(note_path, 'r') as f:
            note_info = json.load(f)
        print("æˆåŠŸè¯»å– note.json æ–‡ä»¶")
    except Exception as e:
        print(f"è¯»å– note.json æ—¶å‡ºé”™: {e}")

# åˆ†ææ ‡ç­¾æ–‡ä»¶ï¼Œæ£€æŸ¥æ˜¯å¦æœ‰é—®é¢˜
print("åˆ†ææ ‡ç­¾æ–‡ä»¶...")
label_stats = {i: 0 for i in range(len(class_names))}
empty_labels = []
problematic_labels = []

for label_file in os.listdir(labels_path):
    if label_file.endswith('.txt'):
        label_path = os.path.join(labels_path, label_file)
        with open(label_path, 'r') as f:
            content = f.read().strip()
            if not content:
                empty_labels.append(label_file)
                continue

            lines = content.split('\n')
            for line in lines:
                parts = line.strip().split()
                if len(parts) != 5:  # YOLOæ ¼å¼åº”è¯¥æœ‰5ä¸ªå€¼ï¼šç±»åˆ«id, x, y, w, h
                    problematic_labels.append(label_file)
                    continue

                try:
                    class_id = int(parts[0])
                    x, y, w, h = map(float, parts[1:5])

                    # æ£€æŸ¥åæ ‡æ˜¯å¦åœ¨[0,1]èŒƒå›´å†…
                    if not (0 <= x <= 1 and 0 <= y <= 1 and 0 <= w <= 1 and 0 <= h <= 1):
                        problematic_labels.append(label_file)
                    else:
                        if class_id in label_stats:
                            label_stats[class_id] += 1
                        else:
                            print(f"è­¦å‘Š: æ ‡ç­¾æ–‡ä»¶ {label_file} ä¸­åŒ…å«æœªçŸ¥ç±»åˆ«ID: {class_id}")
                except ValueError:
                    problematic_labels.append(label_file)

print(f"ç±»åˆ«åˆ†å¸ƒç»Ÿè®¡: {label_stats}")
if empty_labels:
    print(f"è­¦å‘Š: å‘ç° {len(empty_labels)} ä¸ªç©ºæ ‡ç­¾æ–‡ä»¶")
if problematic_labels:
    print(f"è­¦å‘Š: å‘ç° {len(problematic_labels)} ä¸ªæœ‰é—®é¢˜çš„æ ‡ç­¾æ–‡ä»¶")

# è·³è¿‡å¯è§†åŒ–éƒ¨åˆ†ï¼Œé¿å…matplotlibé”™è¯¯

# åˆ›å»ºè®­ç»ƒé›†å’ŒéªŒè¯é›†ç›®å½•ç»“æ„
train_dir = os.path.join(PROJECT_NAME, 'train')
val_dir = os.path.join(PROJECT_NAME, 'val')

train_images_dir = os.path.join(train_dir, 'images')
train_labels_dir = os.path.join(train_dir, 'labels')
val_images_dir = os.path.join(val_dir, 'images')
val_labels_dir = os.path.join(val_dir, 'labels')

os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# è·å–æ‰€æœ‰å›¾åƒæ–‡ä»¶å’Œå¯¹åº”çš„æ ‡ç­¾æ–‡ä»¶
valid_pairs = []
for img_file in os.listdir(images_path):
    if img_file.endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        label_file = os.path.splitext(img_file)[0] + '.txt'
        label_path = os.path.join(labels_path, label_file)

        # åªä½¿ç”¨æœ‰æ•ˆçš„å›¾åƒ-æ ‡ç­¾å¯¹
        print(img_file)
        print(label_file)
        if os.path.exists(label_path) and label_file not in empty_labels and label_file not in problematic_labels:
            valid_pairs.append((img_file, label_file))

if not valid_pairs:
    raise ValueError("æ²¡æœ‰æ‰¾åˆ°æœ‰æ•ˆçš„å›¾åƒ-æ ‡ç­¾å¯¹ï¼Œè¯·æ£€æŸ¥æ•°æ®é›†")

print(f"æ‰¾åˆ° {len(valid_pairs)} ä¸ªæœ‰æ•ˆçš„å›¾åƒ-æ ‡ç­¾å¯¹")

# ç¡®ä¿æ¯ä¸ªç±»åˆ«éƒ½åœ¨è®­ç»ƒé›†å’ŒéªŒè¯é›†ä¸­æœ‰ä»£è¡¨
# æŒ‰ç±»åˆ«åˆ†ç»„
class_samples = {i: [] for i in range(len(class_names))}

for img_file, label_file in valid_pairs:
    label_path = os.path.join(labels_path, label_file)
    with open(label_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if parts:
                class_id = int(parts[0])
                if class_id < len(class_names):
                    class_samples[class_id].append((img_file, label_file))
                    break  # ä¸€ä¸ªæ ·æœ¬åªåˆ†é…ç»™ä¸€ä¸ªç±»åˆ«ï¼ˆå–ç¬¬ä¸€ä¸ªå‡ºç°çš„ç±»åˆ«ï¼‰

# åˆ†å±‚æŠ½æ ·
train_pairs = []
val_pairs = []

for class_id, samples in class_samples.items():
    if samples:  # å¦‚æœè¯¥ç±»åˆ«æœ‰æ ·æœ¬
        random.shuffle(samples)
        split_idx = int(len(samples) * TRAIN_RATIO)
        train_pairs.extend(samples[:split_idx])
        val_pairs.extend(samples[split_idx:])

# å¦‚æœæŸäº›æ ·æœ¬æ²¡æœ‰è¢«åˆ†é…ï¼ˆå¯èƒ½æ˜¯å› ä¸ºæ ‡ç­¾æ–‡ä»¶ä¸ºç©ºï¼‰ï¼Œéšæœºåˆ†é…å®ƒä»¬
unassigned_pairs = [pair for pair in valid_pairs if pair not in train_pairs and pair not in val_pairs]
random.shuffle(unassigned_pairs)
split_idx = int(len(unassigned_pairs) * TRAIN_RATIO)
train_pairs.extend(unassigned_pairs[:split_idx])
val_pairs.extend(unassigned_pairs[split_idx:])

print(f"åˆ†é… {len(train_pairs)} ä¸ªæ ·æœ¬åˆ°è®­ç»ƒé›†")
print(f"åˆ†é… {len(val_pairs)} ä¸ªæ ·æœ¬åˆ°éªŒè¯é›†")

# å¤åˆ¶è®­ç»ƒé›†æ–‡ä»¶
for img_file, label_file in train_pairs:
    # å¤åˆ¶å›¾åƒ
    shutil.copy(
        os.path.join(images_path, img_file),
        os.path.join(train_images_dir, img_file)
    )

    # å¤åˆ¶æ ‡ç­¾æ–‡ä»¶
    shutil.copy(
        os.path.join(labels_path, label_file),
        os.path.join(train_labels_dir, label_file)
    )

# å¤åˆ¶éªŒè¯é›†æ–‡ä»¶
for img_file, label_file in val_pairs:
    # å¤åˆ¶å›¾åƒ
    shutil.copy(
        os.path.join(images_path, img_file),
        os.path.join(val_images_dir, img_file)
    )

    # å¤åˆ¶æ ‡ç­¾æ–‡ä»¶
    shutil.copy(
        os.path.join(labels_path, label_file),
        os.path.join(val_labels_dir, label_file)
    )

# åˆ›å»ºæ•°æ®é›†é…ç½®æ–‡ä»¶
dataset_config = {
    'path': os.path.abspath(PROJECT_NAME),  # ä½¿ç”¨ç»å¯¹è·¯å¾„
    'train': 'train/images',
    'val': 'val/images',
    'names': {i: name for i, name in enumerate(class_names)},
    'nc': len(class_names)  # æ˜ç¡®æŒ‡å®šç±»åˆ«æ•°é‡
}

# ä¿å­˜æ•°æ®é›†é…ç½®æ–‡ä»¶
config_path = os.path.join(PROJECT_NAME, 'dataset_config.yaml')
with open(config_path, 'w') as f:
    yaml.dump(dataset_config, f)

print(f"æ•°æ®é›†é…ç½®æ–‡ä»¶å·²ä¿å­˜åˆ° {config_path}")

# åŠ è½½é¢„è®­ç»ƒæ¨¡å‹
model = YOLO(MODEL_TYPE)

# è®¾ç½®è®­ç»ƒå‚æ•° - å¢å¼ºæ•°æ®å¢å¼ºé€‰é¡¹
train_args = {
    'data': config_path,
    'epochs': EPOCHS,
    'batch': BATCH_SIZE,
    'imgsz': IMAGE_SIZE,
    'project': PROJECT_NAME,
    'name': 'train_results',
    'verbose': True,
    'device': '0',  # ä½¿ç”¨ GPU
    'patience': 30,  # å¢åŠ æ—©åœè€å¿ƒå€¼
    'lr0': 0.01,  # åˆå§‹å­¦ä¹ ç‡
    'lrf': 0.001,  # é™ä½æœ€ç»ˆå­¦ä¹ ç‡
    'momentum': 0.937,  # SGDåŠ¨é‡
    'weight_decay': 0.0005,  # æƒé‡è¡°å‡
    'warmup_epochs': 5,  # å¢åŠ é¢„çƒ­è½®æ•°
    'warmup_momentum': 0.8,
    'warmup_bias_lr': 0.1,
    'box': 7.5,  # è¾¹ç•Œæ¡†æŸå¤±æƒé‡
    'cls': 0.5,  # åˆ†ç±»æŸå¤±æƒé‡
    'dfl': 1.5,  # DFLæŸå¤±æƒé‡
    'hsv_h': 0.015,  # è‰²è°ƒå¢å¼º
    'hsv_s': 0.7,  # é¥±å’Œåº¦å¢å¼º
    'hsv_v': 0.4,  # äº®åº¦å¢å¼º
    'degrees': 0.2,  # å¢åŠ æ—‹è½¬å¢å¼º
    'translate': 0.2,  # å¢åŠ å¹³ç§»å¢å¼º
    'scale': 0.9,  # å¢åŠ ç¼©æ”¾å¢å¼º
    'shear': 0.2,  # å¢åŠ å‰ªåˆ‡å¢å¼º
    'perspective': 0.0005,  # å¢åŠ é€è§†å¢å¼º
    'flipud': 0.5,  # å¢åŠ ä¸Šä¸‹ç¿»è½¬æ¦‚ç‡
    'fliplr': 0.5,  # å·¦å³ç¿»è½¬æ¦‚ç‡
    'mosaic': 1.0,  # é©¬èµ›å…‹å¢å¼º
    'mixup': 0.3,  # å¢åŠ mixupå¢å¼º
    'copy_paste': 0.3,  # å¢åŠ å¤åˆ¶ç²˜è´´å¢å¼º
}

# è®­ç»ƒæ¨¡å‹
try:
    results = model.train(**train_args)
    print("è®­ç»ƒå®Œæˆï¼")
except Exception as e:
    print(f"è®­ç»ƒè¿‡ç¨‹ä¸­å‡ºé”™: {e}")
    # å°è¯•ä½¿ç”¨æ›´ç®€å•çš„å‚æ•°é‡æ–°è®­ç»ƒ
    print("å°è¯•ä½¿ç”¨ç®€åŒ–å‚æ•°é‡æ–°è®­ç»ƒ...")
    simple_train_args = {
        'data': config_path,
        'epochs': EPOCHS,
        'imgsz': IMAGE_SIZE,
        'project': PROJECT_NAME,
        'name': 'train_results',
        # æ·»åŠ åŸºæœ¬çš„æ•°æ®å¢å¼º
        'fliplr': 0.5,  # å·¦å³ç¿»è½¬
        'mosaic': 1.0,   # é©¬èµ›å…‹å¢å¼º
        'hsv_h': 0.015,  # è‰²è°ƒå¢å¼º
        'hsv_s': 0.7,    # é¥±å’Œåº¦å¢å¼º
        'hsv_v': 0.4,    # äº®åº¦å¢å¼º
    }
    try:
        results = model.train(**simple_train_args)
        print("ä½¿ç”¨ç®€åŒ–å‚æ•°è®­ç»ƒå®Œæˆï¼")
    except Exception as e2:
        print(f"ä½¿ç”¨ç®€åŒ–å‚æ•°è®­ç»ƒä»ç„¶å¤±è´¥: {e2}")
        print("è¯·æ£€æŸ¥æ•°æ®é›†æ ¼å¼å’Œå‚æ•°è®¾ç½®")

# éªŒè¯æ¨¡å‹
try:
    best_model_path = os.path.join(PROJECT_NAME, 'train_results', 'weights', 'best.pt')
    if os.path.exists(best_model_path):
        best_model = YOLO(best_model_path)
        print("å·²åŠ è½½æœ€ä½³æ¨¡å‹è¿›è¡ŒéªŒè¯")
    else:
        best_model = model
        print("æœªæ‰¾åˆ°æœ€ä½³æ¨¡å‹ï¼Œä½¿ç”¨æœ€åä¸€ä¸ªæ¨¡å‹è¿›è¡ŒéªŒè¯")

    val_results = best_model.val(data=config_path, conf=CONF_THRESHOLD, iou=0.45)
    print("éªŒè¯ç»“æœ:", val_results)
except Exception as e:
    print(f"éªŒè¯è¿‡ç¨‹ä¸­å‡ºé”™: {e}")

# å¯è§†åŒ–ä¸€äº›é¢„æµ‹ç»“æœ
print("é¢„æµ‹éªŒè¯é›†å›¾åƒ...")
try:
    test_images = [os.path.join(val_images_dir, img) for img in os.listdir(val_images_dir)[:5]]
    if test_images:
        for img_path in test_images:
            # ä½¿ç”¨ä¸åŒçš„ç½®ä¿¡åº¦é˜ˆå€¼è¿›è¡Œé¢„æµ‹
            for conf in [0.01, 0.05, 0.1]:
                try:
                    results = best_model.predict(
                        img_path,
                        save=True,
                        project=PROJECT_NAME,
                        name=f'predictions_conf{conf}',
                        conf=conf,
                        iou=0.45
                    )
                    saved_path = results[0].save(os.path.join(PROJECT_NAME, f'predictions_conf{conf}', os.path.basename(img_path)))
                    print(f"ä½¿ç”¨ç½®ä¿¡åº¦é˜ˆå€¼ {conf} çš„é¢„æµ‹ç»“æœå·²ä¿å­˜åˆ°: {saved_path}")
                    display(Image(filename=saved_path))
                except Exception as e:
                    print(f"é¢„æµ‹å›¾åƒ {img_path} æ—¶å‡ºé”™ (conf={conf}): {e}")
except Exception as e:
    print(f"é¢„æµ‹è¿‡ç¨‹ä¸­å‡ºé”™: {e}")

# å°è¯•åœ¨åŸå§‹å›¾åƒä¸Šè¿›è¡Œé¢„æµ‹
print("åœ¨åŸå§‹å›¾åƒä¸Šè¿›è¡Œé¢„æµ‹...")
try:
    original_images = [os.path.join(images_path, img) for img in os.listdir(images_path)[:5] if img.endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
    if original_images:
        for img_path in original_images:
            try:
                results = best_model.predict(
                    img_path,
                    save=True,
                    project=PROJECT_NAME,
                    name='predictions_original',
                    conf=CONF_THRESHOLD,
                    iou=0.45
                )
                saved_path = results[0].save(os.path.join(PROJECT_NAME, 'predictions_original', os.path.basename(img_path)))
                print(f"åŸå§‹å›¾åƒé¢„æµ‹ç»“æœå·²ä¿å­˜åˆ°: {saved_path}")
                display(Image(filename=saved_path))
            except Exception as e:
                print(f"é¢„æµ‹åŸå§‹å›¾åƒ {img_path} æ—¶å‡ºé”™: {e}")
except Exception as e:
    print(f"åŸå§‹å›¾åƒé¢„æµ‹è¿‡ç¨‹ä¸­å‡ºé”™: {e}")

print("å…¨éƒ¨å®Œæˆï¼")
```
